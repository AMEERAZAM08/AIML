# Q.1 Import the libraries




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# Q.2 Get the Dataset

```python
data = pd.read_csv('Data+set-tips-+Practice+Exercise+Week-2.csv', index_col=0)
data.head()
```

```python
data.info()
```

# Q3. Plot the distribution of total_bill in the given dataset.

```python
sns.distplot(data["total_bill"], kde=False)
plt.show()
```

# Q.4 - Plot the destribution plot for tips.

```python
sns.distplot(data["tip"], kde=False)
plt.show()
```

# Q5. Make a joint plot between tip and total bill.

```python
sns.jointplot(x="tip", y="total_bill", data=data, kind='scatter')
plt.show()
```

# Q6. Give additional attribute kind = "hex" in the above plot.

```python
sns.jointplot(x="tip", y="total_bill", data=data, kind='hex')
plt.show()
```

# Q7. Make pairplot between total bill and tip.

```python
sns.pairplot(data[["tip", "total_bill"]])
plt.show()
```

# Q.8 -  Make striplot between sex and tip.

```python
sns.stripplot(data["sex"], data["tip"])
plt.show()
```

# Q.9 - What other plots you can make for tip and sex variable.Plot them.

```python
sns.swarmplot(data["sex"], data["tip"], hue=data["day"], )
plt.show()
sns.swarmplot(data["sex"], data["tip"], hue=data["smoker"])
plt.show()
sns.swarmplot(data["sex"], data["tip"], hue=data["time"])
plt.show()
```

# Q.10 Make bar plot with day and total bill.

```python
sns.barplot(data["day"], data["total_bill"])
plt.show()
```

# Q.11 - Make a plot to count total males and females for each day.

```python
sns.countplot(data["day"], hue=data["sex"])
plt.show()
```

# Q.12 -  Use factorplot to plot multiple categorical variables like day, smoker, peoplecount.

```python
sns.catplot(x='smoker', y='size',data=data, hue='day', kind='swarm')
plt.show()
```

# Q.13 - Use lmplot() to plot the relation between total bill and tip.


```python
sns.lmplot(x="total_bill", y="tip", data=data)
plt.show()
```

# Q.14 - Plot the above graph for male and female.

```python
sns.lmplot(x="total_bill", y="tip", data=data, hue="sex")
plt.show()
```

# Q.14 - Plot the Q.13 graph for smoker and non-smoker.

```python
sns.lmplot(x="total_bill", y="tip", data=data, hue="smoker", col="sex")
plt.show()
```
